## IR15250 Single Extraction Spectra: Central Bulge

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import astropy.io.fits as fits
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

Read in data

```python
data = fits.getdata("./../creta_output/extractions/IR15250/IR15250_SingleExt_r0.3as.fits")
wavelengths = data[0][0] * (1 + 0.055206)
flux = data[0][5]
flux_err = data[0][6]

"""plt.rcParams.update({
    "lines.color": "white",
    "patch.edgecolor": "white",
    "text.color": "white",
    "axes.facecolor": "black",
    "axes.edgecolor": "lightgray",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "white",
    "figure.facecolor": "black",
    "figure.edgecolor": "black",
    "savefig.facecolor": "black",
    "savefig.edgecolor": "black"})"""
plt.style.use('dark_background')


fig, ax = plt.subplots()
#fig.set_size_inches(12, 12)
ax.errorbar(wavelengths, flux, flux_err, color="cyan")
ax.set_xlabel(r"Wavelength ($\mu$m)")
ax.set_ylabel(r"Flux ($Jy$)")
ax.set_yscale("log")
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.set_title("IR15250 Center Extraction (0.3 arcsec)")
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)
plt.savefig("test_extraction.png", dpi=1000)
```
